```python
import pandas as pd
billboard = pd.read_csv("billboard.csv")
billboard
```

```python
billboard.shape
```

```python
billboard.columns
```

```python
billboard = billboard.dropna(thresh=9)
```

```python
billboard.shape
```

```python
new_billboard = pd.melt(billboard, id_vars=['artiste', 'track', 'genre', 'date.entered', 'date.peaked',],var_name='week',value_name='rank')
new_billboard

```

```python
new_billboard.dtypes
```

```python
new_billboard['week'] = new_billboard['week'].str.extract('(\d+)').astype(int)
```

```python
new_billboard
```

```python
new_billboard.dtypes
```

```python
new_billboard['date.entered'] = pd.to_datetime(new_billboard['date.entered'])
new_billboard['date.peaked'] = pd.to_datetime(new_billboard['date.peaked'])
new_billboard
```

```python
new_billboard['entered_to_peak']=new_billboard['date.peaked']-new_billboard['date.entered']
```

```python
new_billboard
```

```python
new_billboard.dtypes
```

```python
new_billboard['entered_to_peak'] =  new_billboard['entered_to_peak'].dt.days
```

```python
new_billboard.dtypes
```

```python
new_billboard[new_billboard['genre'] == 'Rock'] 
```

```python
new_billboard = new_billboard.sort_values(['rank'])
```

```python
new_billboard
```

```python
new_billboard[new_billboard['entered_to_peak'] == 35] 
```
